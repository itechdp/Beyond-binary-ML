# Household Power Consumption
- Dataset Information

- This archive contains 2075259 measurements gathered in a house located in Sceaux (7km of Paris, France) between December 2006 and November 2010 (47 months).
Notes: 
1.(global_active_power*1000/60 - sub_metering_1 - sub_metering_2 - sub_metering_3) represents the active energy consumed every minute (in watt hour) in the household by electrical equipment not measured in sub-meterings 1, 2 and 3.
2.The dataset contains some missing values in the measurements (nearly 1,25% of the rows). All calendar timestamps are present in the dataset but for some timestamps, the measurement values are missing: a missing value is represented by the absence of value between two consecutive semi-colon attribute separators. For instance, the dataset shows missing values on April 28, 2007.

- Variable Information
1. date: Date in format dd/mm/yyyy
2. time: time in format hh:mm:ss
3. global_active_power: household global minute-averaged active power (in kilowatt)
4. global_reactive_power: household global minute-averaged reactive power (in kilowatt)
5. voltage: minute-averaged voltage (in volt)
6. global_intensity: household global minute-averaged current intensity (in ampere)
7. sub_metering_1: energy sub-metering No. 1 (in watt-hour of active energy). It corresponds to the kitchen, containing mainly a dishwasher, an oven and a microwave (hot plates are not electric but gas powered).
8. sub_metering_2: energy sub-metering No. 2 (in watt-hour of active energy). It corresponds to the laundry room, containing a washing-machine, a tumble-drier, a refrigerator and a light.
9. sub_metering_3: energy sub-metering No. 3 (in watt-hour of active energy). It corresponds to an electric water-heater and an air-conditioner.

### Encoding Date & Time

```python
def encoding_date_time(date,time):
    print(df[date].head())
    print(df[time].head())

    df[date] = pd.to_datetime(df[date], format='%d/%m/%Y')
    df[time] = pd.to_datetime(df[time], format='%H:%M:%S')
    df['Year'] = df[date].dt.year
    df['Month'] = df[date].dt.month
    df['Date'] = df[date].dt.day
    df['Hour'] = df[time].dt.hour
    df['Minute'] = df[time].dt.minute
    df['Second'] = df[time].dt.second

encoding_date_time('Date','Time')
```

```python
df.info()
```

### Converting Feature and label

```python
df_features = df.drop(['Time','Voltage'],axis=1)
df_label = df[['Voltage']]
```

### Checking for an outliers

```python
plt.figure(figsize=(25,10))
df_features.boxplot()
plt.show()
```

### Removing outliers

```python
Q1 = df_features.quantile(0.25)
Q3 = df_features.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR 
upper_bound = Q3 + 1.5 * IQR

df_features_no_outliers = df_features[(df_features>=lower_bound) & (df_features<=upper_bound)]
df_features_no_outliers = pd.DataFrame(df_features_no_outliers,columns=df_features.columns)
plt.figure(figsize=(25,10))
df_features_no_outliers.boxplot()
plt.show()
```

### Filling Missng Values

```python
imputer = SimpleImputer()
df_features_noNaN = imputer.fit_transform(df_features_no_outliers)
df_label_noNaN = imputer.fit_transform(df_label)
df_features_noNaN = pd.DataFrame(df_features_noNaN,columns=df_features.columns)
df_label_nonan = pd.DataFrame(df_label_noNaN,columns=df_label.columns)
```

### Standardization

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_features_noNaN)
df_scaled = pd.DataFrame(df_scaled,columns=df_features.columns)
```

```python
df_features_prepared = df_scaled.copy()
df_label_prepared = pd.DataFrame(df_label_noNaN,columns=df_label_nonan.columns)
```

### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df_features_prepared,df_label_prepared,test_size=0.2,random_state=42)
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
print(X_train.isnull().sum())
print(y_train.isnull().sum())
```

## Model Buildings
- Model Selection 
- Model Evaluation
    - Cross Validation Score
    - Cross Validation Predict  
    - Checking for the Scores
- Model Hyperparameter Tuning
- Model on Test set

### Model Selection

```python
lin_reg = LinearRegression()
tree_reg = DecisionTreeRegressor(min_samples_split=4,min_samples_leaf=2)
elasticnet = ElasticNet(alpha=0.01)
```

### Model Evaluation

### Cross Validation Score

```python
lin_reg_score = cross_val_score(estimator=lin_reg,X=X_train,y=y_train,cv=3,scoring='neg_mean_squared_error')
tree_reg_score = cross_val_score(tree_reg,X_train,y_train,cv=3,scoring='neg_mean_squared_error')
elasticnet_score = cross_val_score(elasticnet,X_train,y_train,cv=3,scoring='neg_mean_squared_error')
```

### Display Cross Validation Score

```python
print(lin_reg_score)
print(tree_reg_score)
print(elasticnet_score)
```

### Cross Validation Predict

```python
lin_reg_pred = cross_val_predict(estimator=lin_reg,X=X_train,y=y_train,cv=3)
tree_reg_pred = cross_val_predict(tree_reg,X_train,y_train,cv=3)
elasticnet_pred = cross_val_predict(elasticnet,X_train,y_train,cv=3)
```

### Display Cross Validation Predict

```python
def display_score(estimator,ytrue,ypred):
    mse = mean_squared_error(ytrue,ypred)
    mae = mean_absolute_error(ytrue,ypred)
    r2 = r2_score(ytrue,ypred)
    print(f"Estimator: {estimator}")
    print(mse)
    print(mae)
    print(r2)

display_score(lin_reg,y_train,lin_reg_pred)
display_score(tree_reg,y_train,tree_reg_pred)
display_score(elasticnet,y_train,elasticnet_pred)
```

### Model Hyperparameter Tuning

```python
param_grid ={
    'criterion': ['squared_error'],
    'min_samples_split': [5, 10],
    'min_samples_leaf': [1, 5]
}
grid_search = GridSearchCV(estimator=tree_reg,param_grid=param_grid,cv=2)
grid_search.fit(X_train,y_train)
print(grid_search.best_score_)
print(grid_search.best_params_)
print(grid_search.best_estimator_)
```

### Fitting Model Test Set

```python
final_model = grid_search.best_estimator_
y_pred = final_model.predict(X_test)
display_score(final_model,ytrue=y_test,ypred=y_pred)
```
